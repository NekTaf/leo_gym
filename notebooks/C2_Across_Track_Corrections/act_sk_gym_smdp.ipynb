```python
import numpy as np
```

```python
from leo_gym.gyms.roe_sk_gym import RoeGym, RoeGymConfig, SatelliteROEConfig
from train_act_sk_cfg import env_cfg

env = RoeGym(cfg=env_cfg, seed=11)
```

```python
from leo_gym.rl_algorithms.ppo_sb3_smdp.ppo_sb3_smdp import SMDP_PPO

agent = SMDP_PPO.load("policies/ppo_smdp_act_sk.zip")
```

```python
try: 
    
    obs, _ = env.reset()
    
    while True:
        
        action,_ = agent.predict(obs)
        obs, rewards, TERMINATED, TRUNCATED, info = env.step(action)
        
        if TRUNCATED or TERMINATED:
            break
                                                        
except KeyboardInterrupt:
    pass

sat1 = env.sat1
```
